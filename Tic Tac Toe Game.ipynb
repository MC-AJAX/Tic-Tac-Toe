```python
#user input function
def userInput(bs,piece): #function that returns user choice
    testList = []           
    for x in bs:
        if bs[x] == " ":   #adds to testList only unused spaces
            testList.append(str(x))  #converts it to a str item
    
    choice = "wrong"
    while choice not in testList: #can't cast to int yet or you get error for floats
        choice = input(f"{piece} choose a space (1-9): ")
        
        if choice not in testList: #informs user of incorrect value and clears cell
            #clear_output()
            print("Please choose a value in the correct range 1-9")
    
    return int(choice)
```

```python
#check if the game is tie
def tieGame(bs,tie):
    emptySpaces = 9
    
    for item in bs:
        if bs[item] == " ":
            emptySpaces -= 1
    
    if emptySpaces == 9:
        return True
    
    return False
```
